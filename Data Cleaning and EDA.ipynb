*****ALGERIAN FOREST FIRE DATASET*****

Data Set Information : 
The dataset has 244 instances of forest fire of two regions in Algeria : Bejala region located in northeast of Algeria and Sidi Bel-abbes region in the nothwest part of the country.

The dataset has 122 instances for both the regions 
The information is from June 2012 to September 2012. It includes 11 attributes and 1 output attribute. The total 244 instances are classified as 138 classes (fire) and 106 classes (no fire)

Attribute Information : 

1. Date : (DD/MM/YYYY) : Weather data information
2. Temp : Temperature in Celcius degress ranging from 22 t0 42
3. RH : Relative Humidity in % 21 t0 90
4. Ws : Wind speed in Km/hr : 6 to 29
5. Rain : Total rain in a day in mm : 0 to 16.8 FWI components 
6. Fine fuel Moisture Code (FFMC) : Index from FWI system : 28.6 to 92.5
7. Duff Moisture Code(DMC) : Index from FWI system : 1.1 to 65.9
8. Initial Speed Index (ISI) : Index from FWI system : 0 to 18.5
9. BuildUp Index (BUI) : 1.1 to 68
10. Fire Weather Index (FWI) : 0 to 31.1
11. Classes : Fire and Not Fire. 

```python
#Library Implementation 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
dataset = pd.read_csv('Algerian_forest_fires_dataset.csv', header = 1 )
```

```python
## Data Cleaning 

## Missing Values 
dataset.isnull().sum()
```

```python
dataset[dataset.isnull().any(axis = 1)]
```

The dataset is converted into two sets based on the region from 123rd index, we can make a new column based on region

1. "Bejaia Region Dataset"
2. "Sidi-Bel Abbes Dataset"

Add new column with region

```python
dataset.loc[:123, "Region"] = 0
dataset.loc[123:, "Region"] = 1
df = dataset
```

```python
df.info()
```

```python
df[['Region']] = df[['Region']].astype(int)
```

```python
df.head()
```

```python
## Removing the null values 
df = df.dropna().reset_index(drop = True)
```

```python
df.isnull().sum()
```

```python
df.iloc[[122]]
```

```python
df = df.drop(122).reset_index(drop = True)
```

```python
df.iloc[[122]]
```

```python
df.columns
```

```python
## Fix spaces and column names 
df.columns = df.columns.str.strip()
df.columns
```

```python
## Change required columns as integer data type
df[['month','day','year','Temperature','RH']] = df[['month','day','year','Temperature','RH']].astype(int)
```

```python
df.info()
```

```python
## Changing other columns to float data type

objects = [features for features in df.columns if df[features].dtypes == 'O']
```

```python
for i in objects:
    if i != 'Classes':
        df[i] = df[i].astype(float)
```

```python
## Saving cleaned dataset 

df.to_csv('Algerian_forest_fires_cleaned_dataset.csv', index = False)
```
